Importamos o arquivo com o parâmetro "*header = None*", uma vez que o arquivo de exemplo não possui cabeçalho

```python
import pandas as pd
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data"
df = pd.read_csv(url, header = None)
```

Exibindo os primeiros registros para verificar se o arquivo foi carregado corretamente. Perceba que o Pandas abribui um nome de coluna automaticamente (0 a 25), uma vez que o nosso arquivo não possui cabeçalho.

```python
df.head()
```

Substituindo os nomes das colunas

```python
headers=["symboling","normalized-losses","make","fuel-type","aspiration","num-of-doors","body-style","drive-wheels","engine-location","wheel-base","length","width","height","curb-weight","engine-type","num-of-cylinders","engine-size","fuel-system","bore","stroke","compression-ratio","horsepower","peak-rpm","city-mpg","highway-mpg","price"]
df.columns = headers
```
